## Problem Statement
To study and compare four major deep learning frameworks—TensorFlow, Keras, Theano, and PyTorch. The aim is to identify their distinct features, capabilities, advantages, limitations, and typical use-cases. This comparison will serve as a foundation for understanding which framework is suitable for different kinds of machine learning and deep learning projects. <p>

## Description of the Problem
Deep learning frameworks provide an ecosystem for designing, training, and deploying neural networks. Each framework has evolved over time with unique design philosophies, computational models, and levels of abstraction.
This assignment requires a detailed study of each package from both theoretical and practical perspectives. A small demonstration using a suitable dataset (MNIST, CIFAR-10, or custom) should be included to show how a simple neural network is implemented in each framework. <p>

## Simplified Explanation of Deep Learning Packages ⭐

1. TensorFlow
Graph Type: Static + Dynamic
Ease of Use: Medium (easy after TF 2.0)
Performance: Very fast, supports GPU + TPU
Popularity: Huge in industry
Unique Features: TensorBoard, TensorFlow Lite, TPU support
Summary: Great for large-scale and production ML systems.

2. Keras
Graph Type: Uses TensorFlow backend
Ease of Use: Very easy (high-level API)
Performance: Fast because it sits on TensorFlow
Popularity: Very popular for learning & prototyping
Unique Features: Sequential/Functional API, clean & simple code
Summary: Best for beginners and quick model building.


4. Theano
Graph Type: Static
Ease of Use: Hard, low-level
Performance: Fast but outdated
Popularity: Not used anymore
Unique Features: First deep learning library (historical)
Summary: Important historically, but no longer used.


6. PyTorch
Graph Type: Dynamic (define-by-run)
Ease of Use: Very easy, Pythonic
Performance: Fast GPU support
Popularity: #1 in research
Unique Features: Flexible, great debugging, HuggingFace support
Summary: Best for research and experimentation. <p>
